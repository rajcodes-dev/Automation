```python
conn = sqlite3.connect('example.db', isolation_level=None)
```

```python
conn.execute('CREATE TABLE IF NOT EXISTS cats (name TEXT NOT NULL,' \
'birthday TEXT,' \
'fur TEXT,' \
'weight_kg REAL) STRICT')
```

```python
conn.execute('INSERT INTO cats VALUES ("Zophie", "22-04-24", "black", 5.6)')
```
