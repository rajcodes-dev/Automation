```python
# rollback
```

```python
import sqlite3
```

```python
conn = sqlite3.connect('sweigartcats.db', isolation_level=None)
```

```python
conn.execute('BEGIN')
```

```python
conn.execute('INSERT INTO cats VALUES ("Socks", "2022-04-04", "white", 4.2)')
```

```python
conn.execute('INSERT INTO cats VALUES ("Fluffy", "2022-10-30", "gray", 4.5)')
```

```python
conn.rollback()
```

```python
conn.execute('SELECT * FROM cats WHERE name = "Socks"').fetchall()
```

```python
conn.execute('SELECT * FROM cats WHERE name = "Fluffy"').fetchall()
```

```python
conn.execute('INSERT INTO cats VALUES ("Socks", "2022-04-04", "white", 4.2)')
```

```python
conn.execute('INSERT INTO cats VALUES ("Fluffy", "2022-10-30", "gray", 4.5)')
```

```python
conn.commit()
```

```python
conn.execute('SELECT * FROM cats WHERE name = "Fluffy"').fetchall()
```

```python
conn.execute('SELECT * FROM cats WHERE name = "Socks"').fetchall()
```

```python
# Creating foreign keys
```

```python
import sqlite3
```

```python
conn = sqlite3.connect('sweigartcats.db', isolation_level= None)
```

```python
conn.execute('PRAGMA foreign_keys = ON')
```

```python
conn.execute('CREATE TABLE IF NOT EXISTS vaccinations (vaccine TEXT, date_administered TEXT, administered_by TEXT, cat_id INTEGER, FOREIGN KEY(cat_id) REFERENCES cats(rowid))')
```

```python
conn.execute("ALTER TABLE cats RENAME TO cats_old")
```

```python
conn.execute(
    """CREATE TABLE cats (
    cat_id INTEGER PRIMARY KEY, 
    name TEXT NOT NULL, 
    birthdate TEXT, 
    fur TEXT, 
    weight_kg REAL
) STRICT"""
)
```

```python
conn.execute(
    """INSERT INTO cats (name, birthdate, fur, weight_kg) 
                SELECT name, birthdate, fur, weight_kg FROM cats_old"""
)
```

```python
conn.execute("DROP TABLE IF EXISTS vaccinations")
conn.execute(
    """CREATE TABLE vaccinations (
    vaccine TEXT,
    date_administered TEXT, 
    administered_by TEXT, 
    cat_id INTEGER,
    FOREIGN KEY(cat_id) REFERENCES cats(cat_id)
) STRICT"""
)
```

```python
conn.execute('INSERT INTO vaccinations VALUES ("rabbies", "2023-06-06", "Dr. Echo", 1)')
```

```python
conn.execute('INSERT INTO vaccinations VALUES ("FeLV", "2023-06-06", "Dr. Echo", 1)')
```
