```python
# rotating a pdf
```

```python
import pypdf
```

```python
writer = pypdf.PdfWriter()
```

```python
writer.append('Recursion_Chapter1.pdf')
```

```python
for i in range(len(writer.pages)):
    writer.pages[i].rotate(90)
```

```python
with open('rotated.pdf', 'wb') as file:
    writer.write(file)
```
