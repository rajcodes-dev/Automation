```python
import re 
```

```python
pattern = re.compile(r'43!?')
```

```python
pattern.search('43!')
```

```python
pattern.search('43')
```

```python
pattern = re.compile(r'43?!')
```

```python
pattern.search('43!')
```

```python
pattern.search('4!')
```

```python
pattern.search('43') == None
```

```python
pattern = re.compile(r'(\d{3}-)?\d{3}-\d{4}')
```

```python
match1 = pattern.search('My phone number is 234-239-2929')
```

```python
match1.group()
```

```python
match2 = pattern.search('My phone number is 239-2929')
```

```python
match2.group()
```

```python
pattern = re.compile('Eggs( and spam)*')
```

```python
pattern.search('Eggs and spam')
```

```python
pattern.search('Eggs and spam and spam and spam')
```

```python
pattern = re.compile('Eggs( and spam)+')
```

```python
pattern.search('Eggs')
```

```python
pattern.search('Eggs and spam')
```

```python
pattern = re.compile(r'(AH){4}')
```

```python
match1 = pattern.search('AHAHAHAH')
```

```python
match1.group()
```

```python
match2 = pattern.search('AHAH')
```

```python
match2 == None
```
