```python
# Greedy and non greedy matching
```

```python
greedy_pattern = re.compile(r'(HA){3,5}') # greedy pattern
```

```python
match = greedy_pattern.search('HAHAHAHAHA')
```

```python
match.group()
```

```python
lazy_pattern = re.compile(r'(HA){3,5}?') # lazy pattern
```

```python
match = lazy_pattern.search('HAHAHAHAHA')
```

```python
match.group()
```

```python
name_pattern = re.compile(r'First name: (.*) Last name: (.*)')
```

```python
name_match.group()
```

```python
name_match.group(1)
```

```python
name_match.group(2)
```

```python
lazy_pattern = re.compile(r'<.*?>')
```

```python
match_pattern.group()
```

```python
greedy_pattern = re.compile(r'<.*>')
```

```python
match_pattern1 = greedy_pattern.search("<To serve man> for dinner.>")
```

```python
no_newline_re = re.compile('.*')
```

```python
no_newline_re.search("Hi, How are you?\nI am fine.\nWhat's about you?\nI am also fine. ").group()
```

```python
newline_re = re.compile('.*', re.DOTALL)
```

```python
newline_re.search(
    "Hi, How are you?\nI am fine.\nWhat's about you?\nI am also fine. "
).group()
```
