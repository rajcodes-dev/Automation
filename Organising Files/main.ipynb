```python
# working with zip
```

```python
import zipfile
```

```python
with open('file1.txt', 'w', encoding='utf-8') as file:
    file.write('test_data' * 10000)
```

```python
with zipfile.ZipFile('example.zip', 'w') as example_zip:
    example_zip.write('file1.txt', compress_type=zipfile.ZIP_DEFLATED, compresslevel=9)
```

```python
# reading zipfiles
```

```python
example_zip = zipfile.ZipFile('example.zip')
```

```python
example_zip.namelist()
```

```python
file_info = example_zip.getinfo('file1.txt')
```

```python
file_info.file_size
```

```python
file_info.compress_size
```

```python
f"Compressed file is {round(file_info.file_size/file_info.compress_size, 2)}x smaller!"
```

```python
example_zip.close()
```
